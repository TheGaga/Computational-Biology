```python
from collections import OrderedDict
from operator import itemgetter

#word_dict will save the occurrances of the word, the key is the word and the value is the occurrances
word_dict = dict()

#read line by line, if word exist, occurance +1, if not create item
with open('romeoandjuliet.txt','r') as f:
    for line in f:
        for word in line.split():
            
            word =  ''.join(filter(str.isalpha, word))
            word = word.lower()
            if word in word_dict:
                word_dict[word]+=1
            else:
                word_dict[word] =1

#sort the dictionary in reverse and save into the order_dict
ordered_dict = OrderedDict(sorted(word_dict.items(), key=itemgetter(1),reverse=True))
print (ordered_dict)
```

```python
import matplotlib.pyplot as plt

#topten save the first 10 items of ordered_dict
topten = dict(list(ordered_dict.items())[:10])

#plot according to the keys and values of topten
plt.bar(range(10), list(topten.values()), align='center')
plt.xticks(range(10), list(topten.keys()));
plt.show()
```
